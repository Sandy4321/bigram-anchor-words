```python
import pandas as pd
import seaborn as sns

%pylab inline
sns.set_style("darkgrid")
sns.set_context(rc={"figsize": (16, 12)})
sns.set(font_scale=1.7)
```

```python
df1 = pd.read_csv('bi.txt', sep=' ', header=None)
df2 = pd.read_csv('noun+anch+plsa.txt', sep=' ', header=None)
anch = 2099.08, 2330.34, 0.94, 0.63, 0.53
label1, label2 = 'bi+an+pl', 'an+pl+no'

f = lambda d: map(lambda x: float(x[:-1]) if x[-1] == ',' else float(x), map(str, d))
dots_x = range(10)
```

```python
i, loc, ti = 0, 4, u'Перплексия' 

pylab.ylabel(ti)
pylab.xlabel(u'Итерации')

st = u'Обучение '

pylab.plot(dots_x, f(df1[i].values), label=st+label1)
pylab.plot(dots_x, f(df2[i].values), label=st+label2) 

i = 1
st = u'Контроль '

pylab.plot(dots_x, f(df1[i].values), label=st+label1)
pylab.plot(dots_x, f(df2[i].values), label=st+label2)

pylab.plot([0], [1300])

pylab.legend(loc=loc)
pylab.savefig('%s.png' % i, dpi=300)
pylab.show()
```

```python
i, loc, ti, sub_ti = 2, 4, u'Когерентность', u' среднее' 

pylab.ylabel(ti)
pylab.xlabel(u'Итерации')

pylab.plot(dots_x, f(df1[i].values), label=label1+sub_ti)
pylab.plot(dots_x, f(df2[i].values), label=label2+sub_ti)

i, sub_ti = 3, u' медиана'

pylab.plot(dots_x, f(df1[i].values), label=label1+sub_ti)
pylab.plot(dots_x, f(df2[i].values), label=label2+sub_ti)

pylab.legend(loc=loc)
pylab.savefig('%s.png' % i, dpi=300)
pylab.show()
```

```python
i, loc, ti = 4, 4, u'Уникальность ядер' 

pylab.ylabel(ti)
pylab.xlabel(u'Итерации')

pylab.plot(dots_x, f(df1[i].values), label=label1)
pylab.plot(dots_x, f(df2[i].values), label=label2)

pylab.legend(loc=loc)
pylab.savefig('%s.png' % i, dpi=300)
pylab.show()
```
